```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X = customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
X.head()
```

```python
y = customers['Yearly Amount Spent']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
prediction = lm.predict(X_test)
```

```python
plt.scatter(y_test,prediction)
plt.xlabel('Y Test(True Values)')
plt.ylabel('Predicted Value')
```

```python
MAE = metrics.mean_absolute_error(y_test,prediction)
MSE = metrics.mean_squared_error(y_test,prediction)
RMSE = metrics.root_mean_squared_error(y_test,prediction)
print(f"MAE: {MAE}\nMSE: {MSE}\nRMSE: {RMSE}")
```

```python
metrics.explained_variance_score(y_test,prediction)
```

```python
sns.histplot((y_test-prediction), kde=True, bins=60)
```

```python
cdf = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
cdf
```
